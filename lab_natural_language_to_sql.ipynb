````python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch
import accelerate
````

````python
model_name = "defog/sqlcoder-7b"
````

We need to create the Quantization configuration to load the Model.

It is a large model and I want it to fit in a 16GB GPU, I'm going to use a 4 bits quantization.

If you want to learn more about quantization, refer to this article: [QLoRA: Training a Large Language Model on a 16GB GPU.](https://medium.com/towards-artificial-intelligence/qlora-training-a-large-language-model-on-a-16gb-gpu-00ea965667c1)

You can try to use this model in a 8 bit quantizations and check in you see any improvements in the results.

````python
bnb_config = BitsAndBytesConfig(
  load_in_4bit=True,
  bnb_4bit_use_double_quant=True,
  bnb_4bit_quant_type="nf4",
  bnb_4bit_compute_dtype=torch.bfloat16
)
````

To load the model I pass to the AutoModelForCasualLM teh quantization configurations, and HuggingFace take care of all the hard work.

````python
foundation_model = AutoModelForCausalLM.from_pretrained(model_name,
                    quantization_config=bnb_config,
                    device_map='auto',
                    use_cache = True)
````

````python
tokenizer = AutoTokenizer.from_pretrained(model_name)
eos_token_id = tokenizer.convert_tokens_to_ids(["```"])[0]
````

This function wraps the call to *model.generate*

````python
#this function returns the outputs from the model received, and inputs.
def get_outputs(model, inputs, max_new_tokens=400):
    outputs = model.generate(
        input_ids=inputs["input_ids"],
        attention_mask=inputs["attention_mask"],
        num_return_sequences=1,
        eos_token_id=eos_token_id,
        pad_token_id=eos_token_id,
        max_new_tokens=max_new_tokens,
        do_sample=False,
        num_beams=5
    )
    return outputs
````

# Prompt without Shots.
In this first PROMPT we are going to give Instructions to the model and pass the structure of the Database.

The instructions are significantly different from those we are passing to GPT-3.5-Turbo. This model is really well fine-tuned, but it is smaller than GPT-3.5.

We need to be more clear with the instructions, as it does not have the same capacity to understand our orders as GPT-3.5.

````python
sp_nl2sql = """
    ### Instructions:
Your task is convert a question into a SQL query, given a SQL database schema.
Adhere to these rules:
- **Deliberately go through the question and database schema word by word** to appropriately answer the question

    ### Input
    Generate a SQL query that answers the question below.
    This query will run on a database whose schema is represented in this string:

    CREATE 3+ TABLES HERE

    ### Response
    Based on your instructions, here is the SQL query I have generated to answer the question
    `{question}`:
    ```sql3
    """
````

````python
sp_nl2sql = sp_nl2sql.format(question="YOUR QUERY HERE")
print(sp_nl2sql)
````

````python
input_sentences = tokenizer(sp_nl2sql, return_tensors="pt").to('cuda')
response = get_outputs(foundation_model, input_sentences, max_new_tokens=400)
SQL = tokenizer.batch_decode(response, skip_special_tokens=True)
````

````python
#Empty the cache in orde to do more calls without problems.
torch.cuda.empty_cache()
````

````python
print(SQL[0].split("```sql3")[-1].split("```")[0].split(";")[0].strip() + ";")
````

The SQL Order is correct.

## Findings Report

### General Observations
During this lab, different prompt formulations were tested to convert natural language questions into SQL queries.  
The structure and clarity of the prompt had a direct impact on the correctness of the generated SQL.

Providing a clear schema and simple question wording generally led to better results.

---

### Variations that did not work well

Some prompt variations produced incorrect or unreliable outputs.

 When the question was too vague, the model sometimes:
  - Assumed relationships that were not explicitly defined
  - Used fields that were not present in the schema
  - Confused tables that had similar purposes

 In more complex questions, the model occasionally:
  - Missed necessary JOIN conditions
  - Used incorrect grouping logic
  - Selected columns unrelated to the requested metric

 When the prompt wording introduced ambiguity (for example linking education and salary indirectly), the model sometimes:
  - Focused on the wrong table
  - Generated logically valid SQL that did not actually answer the question

---

### Hallucinations / Logical Errors

Typical issues observed:

 Use of non-existing columns  
 Incorrect aggregation logic (missing GROUP BY)  
 Wrong table connections  
 Returning technically valid SQL that did not match the intent of the question  

These errors appeared more frequently when:

- The prompt did not clearly specify what needed to be compared
- The relationship between tables had to be inferred
- The question required interpretation instead of direct retrieval

---

### What improved performance

The following helped generate better SQL:

 Clear schema visibility  
 Direct and simple question phrasing  
 Explicit comparison goals (e.g. highest salary)  
 Avoiding unnecessary narrative in the prompt  

Prompts that were shorter and focused on a measurable objective produced more reliable queries.

---

### What I learned

 Prompt clarity strongly affects SQL accuracy  
 Models tend to infer missing structure when instructions are ambiguous  
 Explicit goals reduce hallucination  
 Simpler prompts often outperform more complex ones  
 Linking concepts (like education and salary) requires careful wording  

Overall, combining:

- clear schema  
- precise question  
- minimal ambiguity  

produces the most consistent Text-to-SQL results.
